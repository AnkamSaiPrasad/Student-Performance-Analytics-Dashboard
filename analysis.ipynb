```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('student_data.csv')
df
```

```python
corr = df[['Marks_Math', 'Marks_English', 'Marks_Science', 'Attendance (%)', 'Logins_Per_Week']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
df_sorted = df.sort_values(by='Average_marks', ascending=False)
plt.figure(figsize=(15,5))
sns.barplot(data=df_sorted, x='Name', y='Average_marks')
plt.title('Student Average Marks')
plt.xticks(rotation=90)
plt.show()
```

```python
df['Performance_Group'] = pd.cut(df['Average_marks'], bins=[0, 50, 75, 100], labels=['Low', 'Medium', 'High'])
sns.scatterplot(data=df, x='Attendance (%)', y='Average_marks', hue='Performance_Group')
plt.title('Attendance vs Average Marks')
plt.show()
```

```python
top_students = df_sorted.head(5)
struggling_students = df_sorted.tail(5)
print("Top Students:\n", top_students[['Name', 'Total_marks', 'Attendance (%)']])
print("\nStruggling Students:\n", struggling_students[['Name', 'Total_marks', 'Attendance (%)']])
```
